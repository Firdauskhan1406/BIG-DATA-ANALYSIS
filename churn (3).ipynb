```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Separate numeric and categorical columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
# Fill missing values in numeric columns using the median
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
# Step 3: Select features for clustering (BounceRates and ExitRates)
df_clustering = df[['BounceRates', 'ExitRates']]
```

```python
df_clustering
```

```python
# Step 4: Normalize the features (scaling)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clustering)
```

```python
# Step 5: Elbow method to find the optimal number of clusters
inertia = []
silhouette_scores = []
K = range(2, 11)  # Checking clusters between 2 and 10

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

    # Calculate silhouette score for each k
    cluster_labels = kmeans.predict(df_scaled)
    silhouette_avg = silhouette_score(df_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)
```

```python
# Step 6: Plotting the Elbow method for inertia
plt.figure(figsize=(8, 6))
plt.plot(K, inertia, 'bo-', marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal K')
plt.show()
```

```python
# Step 7: Plotting the Silhouette Scores for each K
plt.figure(figsize=(8, 6))
plt.plot(K, silhouette_scores, 'go-', marker='x')
plt.xlabel('Number of clusters')
plt.ylabel('Elbow Method ')
plt.title('Elbow Method  For Different K')
plt.show()
```

```python
# Step 8: Based on the elbow and silhouette, assume optimal K is 4 (just as an example, adjust based on plot)
optimal_k = 4
```

```python
# Step 9: Apply KMeans with the optimal number of clusters
kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans_optimal.fit_predict(df_scaled)
```

```python
# Step 10: Visualizing the clustering
plt.figure(figsize=(8, 6))
plt.scatter(df_clustering['BounceRates'], df_clustering['ExitRates'], c=df['Cluster'], cmap='viridis', marker='o')
plt.xlabel('Bounce Rates')
plt.ylabel('Exit Rates')
plt.title(f'Customer Segmentation using K-Means (K={optimal_k})')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Step 11: Analyzing the cluster assignments
print(f'Inertia for k={optimal_k}: {kmeans_optimal.inertia_}')
print(f'Silhouette Score for k={optimal_k}: {silhouette_score(df_scaled, df["Cluster"])}')
print(df['Cluster'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Revenue is in True/False format: True = made a purchase (not churned), False = no purchase (churned)
df['Churn'] = df['Revenue'].apply(lambda x: 1 if not x else 0)  # 1 = churn (no purchase), 0 = no churn (purchase)
```

```python
df
```

```python
# Step 3: Prepare features (X) and target variable (y)
# Select relevant features for prediction
features = [
    'BounceRates',
    'ExitRates',
    'Administrative_Duration',
    'Informational_Duration',
    'ProductRelated_Duration',
    'PageValues',
    'VisitorType',
    'TrafficType',
    'Weekend',
    'SpecialDay'
]
X = df[features]
y = df['Churn']  # Target variable: 1 for churned, 0 for not churned
```

```python
# Step 4: Preprocess the data
# Identify categorical and numerical features
categorical_features = ['VisitorType', 'TrafficType']
numerical_features = [col for col in features if col not in categorical_features]
```

```python
# Create a preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ]
)
```

```python
# Step 5: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 6: Create a pipeline with preprocessing and model
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])
```

```python
# Step 7: Train the model
pipeline.fit(X_train, y_train)
```

```python
# Step 8: Make predictions on the test set
y_pred = pipeline.predict(X_test)
y_pred_prob = pipeline.predict_proba(X_test)[:, 1]  # Probability of positive class
```

```python
# Step 9: Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print(f"ROC AUC Score: {roc_auc_score(y_test, y_pred_prob)}")
```

```python
# Optional: Analyze feature importance (coefficients)
logreg = pipeline.named_steps['classifier']
coefficients = pd.DataFrame(logreg.coef_[0], index=preprocessor.named_transformers_['num'].get_feature_names_out(numerical_features).tolist() + list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)), columns=['Coefficient'])
print("\nFeature Importance:")
print(coefficients)
```
